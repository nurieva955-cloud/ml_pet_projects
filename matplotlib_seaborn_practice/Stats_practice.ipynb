### Ящик с усами

#### Здесь границы ящика это Q1 (25%) = 26 (нижняя граница), Q3 (75%) = 78 (верхняя граница). IQR = 78 - 26 = 52
#### Q2 (50% - медиана) = 52. По определению усы доходят до Q1 - 1.5 * IQR (внизу) и Q3 + 1.5 * IQR (наверху)
#### НО на графике они не рисуются по интервалам, а только по фактическим значениям. Интервалы здесь используются для определения выбросов.
#### Причем эти интервалы не включаются. То есть если какие то значения входят в границы усов, то они будут использоваться для отрисовки.
#### Всё что больше интервалов - выбросы. Например, у нас:
#### Верхняя граница усов = 78 + 1.5 * 52 = 156 (то есть все значения меньше 156 попадут в ус)
#### Нижняя граница усов  = 26 - 1.5 * 52 = -52 (то есть все значения больше -52 попадут в ус)
#### Например, если указать в списке дополнительным элементом значение 150, то ус дотянется до 150. Но если будет 156 и больше, это уже выброс.

```python
list_for_boxplot = [i for i in range(1,104)]+[155]
plt.boxplot(list_for_boxplot) 
print(list_for_boxplot)
print(f'Длина выборки (списка) = {len(list_for_boxplot)}')
plt.title('Ящик с усами')
plt.show()
```
